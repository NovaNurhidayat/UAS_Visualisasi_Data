```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
kebutuhan_per_bulan_2 = pd.read_csv('kebutuhan_per_bulan_2.csv')
```

```python
kebutuhan_per_bulan_2.head()
```

SCATTER PLOT 1 KEBUTUHAN PERTAHUN BERDASARKAN SPESIFIK PROVINSI BANTEN

```python
df = pd.DataFrame(kebutuhan_per_bulan_2 )
# scatter plot 1 hanya menampikan kebutuhan tahunan dari 1 provinsi
spesifik_provinsi = 'BANTEN'
df_spesifik_provinsi = df.loc[df['Propinsi'] == spesifik_provinsi]

x = df_spesifik_provinsi['tahun']
y = df_spesifik_provinsi['kebutuhan']

plt.title(f'PLOT 1 - STATISTIK KEBUTUHAN {spesifik_provinsi}')
plt.xlabel("Tahun")
plt.ylabel("Kebutuhan")

plt.scatter(x, y, color='#D04', marker='s')

plt.xlim(min(x), max(x))
# Menampilkan data provinsi tertentu

plt.show()
```

```python
profile_pelanggan = pd.read_csv('profile_pelanggan.csv', names=['id', 'provinsi'])
```

```python
profile_pelanggan.head()
```

SCATER PLOT 2 MENENTUKAN JUMLAH PELANGGAN TERBANYAK BERDASARKAN PROVINSI

```python
province_counts = profile_pelanggan['provinsi'].value_counts().reset_index()
province_counts.columns = ['provinsi', 'count']

plt.scatter(province_counts['provinsi'], province_counts['count'])

plt.title('PLOT 2 - jumlah pelangggan terbanyak berdasarkan provinsi ')
plt.xlabel('Provinsi')
plt.ylabel('Jumlah pelanggan')

plt.xticks(rotation=90)

plt.show()
```

SCATER PLOT 3 - JUMLAH PESANAN SETIAP KOTA

```python
x = supply_kantor['Kota']
y = supply_kantor['Jumlah']
plt.figure(figsize=(10, 8))
plt.scatter(x, y)

plt.xlabel('Kota')
plt.ylabel('Jumlah')

plt.title('Plot 3 - Jumlah Pesanan dari Setiap Kota')

plt.show()
```

```python
transaksi = pd.read_csv('transaksi2.csv', names=['date', 'id_product', 'quantity', 'id_penjualan'])
```

```python
transaksi.head()
```

SCATER PLOT 4 - MENENTUKAN KUANTITAS TERBANYAK BERDASARKAN TANGGAL

```python
transaksi['date'] = pd.to_datetime(transaksi['date'])

plt.scatter(transaksi['date'], transaksi['quantity'])

plt.title('Plot 4 - KUANTITAS TERBANYAK BERDASARKAN TANGGAL')
plt.xlabel('Date')
plt.ylabel('Quantity')

plt.show()
```

SCATER PLOT 5 - Menentukan jumlah produk terbanyak

```python
product_quantities = transaksi.groupby('id_product')['quantity'].sum().reset_index()

plt.scatter(product_quantities['id_product'], product_quantities['quantity'])

plt.title('Plot 5 - Jumlah Produk terbanyak')
plt.xlabel('Product ID')
plt.ylabel('Total Quantity')

plt.show()
```
